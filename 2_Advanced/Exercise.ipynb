```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#Filter the dataframe for data nalyst in the United states only

df_DA_US = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] == "United States")].copy()

#Dropping the null values in job_skills colum

df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])

df_DA_US
```

```python
df_DA_US = df_DA_US.explode("job_skills")
df_DA_US[["job_skills", "salary_year_avg"]].head(5)

```

```python
top_pay  = df_DA_US.groupby("job_skills")["salary_year_avg"].agg(["count", "median"]).sort_values(by="median", ascending=False)
top_pay = top_pay.head(10)
top_pay
```

```python
high_demand = df_DA_US.groupby("job_skills")["salary_year_avg"].agg(["count", "median"]).sort_values(by="count", ascending=False)
high_demand = high_demand.head(10).sort_values(by="median", ascending=False)
high_demand
```

```python
fig, ax = plt.subplots(2,1)

top_pay[::-1].plot(kind='barh', ax= ax[0], y = "median", legend = False)
ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')
ax[0].set_ylabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

high_demand[::-1].plot(kind='barh', ax= ax[1], y = "median", legend = False)
ax[1].set_title('Top 10 Highest Demand Skills for Data Analysts')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary (USD)')

plt.tight_layout()

```

**Seaborn Implementation**

```python
import seaborn as sns
fig, ax = plt.subplots(2,1)

sns.set_theme(style="ticks")


sns.barplot(data= top_pay, x = "median", y = top_pay.index, hue = "median", ax= ax[0], palette= "dark:b_r")

ax[0].legend().remove()
# top_pay[::-1].plot(kind='barh', ax= ax[0], y = "median", legend = False)
ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')
ax[0].set_ylabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

# high_demand[::-1].plot(kind='barh', ax= ax[1], y = "median", legend = False)

sns.barplot(data=high_demand, x = "median", y = high_demand.index, hue = "median", ax=ax[1], palette= "light:b")
ax[1].legend().remove()
ax[1].set_title('Top 10 Highest Demand Skills for Data Analysts')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary (USD)')

plt.tight_layout()
```
