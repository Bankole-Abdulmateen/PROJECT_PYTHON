```python
#Importing Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data cleanup

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])

import ast

df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

```

```python
#Using the explode function to split the skills in a list to generate individual rows

df_exploded = df.explode("job_skills")
df_exploded.head()
```

```python
df_exploded ["job_skills"].value_counts().head(10).plot(kind="bar")
```

```python
skills_count = df_exploded.groupby(["job_title_short", "job_skills"]).size()
```

```python
df_skills_count = skills_count.reset_index(name = "skill_count")
df_skills_count
```

```python
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True) 
df_skills_count
```

```python
skill = "Data Analyst"
top_skills = 10

df_skill_final = df_skills_count[df_skills_count["job_title_short"] == skill].head(top_skills)

df_skill_final
```

```python
df_skill_final.plot(kind='barh', x = "job_skills", y = "skill_count")
plt.gca().invert_yaxis()
```

```python
df_DA = df[df["job_title_short"] == "Data Analyst"].copy()
df_DA
```

```python
df_DA["job_posted_month_no"] = df_DA["job_posted_date"].dt.month
df_DA
```

```python
df_DA_exploded = df_DA.explode("job_skills")
```

```python
df_DA_pivot = df_DA_exploded.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)

df_DA_pivot
```

```python
df_DA_pivot.loc["Total"] = df_DA_pivot.sum()

df_DA_pivot = df_DA_pivot[df_DA_pivot.loc["Total"].sort_values(ascending=False).index]

df_DA_pivot = df_DA_pivot.drop("Total")
df_DA_pivot
```

```python
# code snippet from `6_Exercise_Job_Demand.ipynb`
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')

df_DA_pivot
```

```python
# Customizing the Line plot

df_DA_pivot.iloc[:, :5].plot(kind="line",
 linewidth = 4,
 linestyle = ":", # options: '-', '--', '-.', ':'
 marker = "o", # options: 'o', 'x', '+', '*', 's', ',', '.', '1', '2', '3', '4', ...
 markersize = 5, 
 colormap= "viridis",
 figsize=(10,3)
 ) 


plt.title('Top 5 Skills for Data Analysts per Month')
plt.grid(True)
plt.ylabel('Count')
plt.xlabel('')
plt.ylim(0,12000)
plt.show()
```
