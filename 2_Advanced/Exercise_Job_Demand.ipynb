```python
#Importing Libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data cleanup

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.head()
```

```python
df_Us = df[df["job_country"] == "United States"].copy()
df_Us.iloc[0:3]
```

```python
df_Us["job_posted_month"] = df_Us["job_posted_date"].dt.strftime('%B')
df_Us
```

```python
df_Us_pivot = df_Us.pivot_table(index ="job_posted_month", columns="job_title_short", aggfunc="size")
```

```python
df_Us_pivot.reset_index(inplace=True)

df_Us_pivot ['month_no'] = pd.to_datetime(df_Us_pivot["job_posted_month"], format="%B").dt.month
df_Us_pivot.sort_values("month_no", inplace=True)
df_Us_pivot.set_index("job_posted_month", inplace=True)
df_Us_pivot.drop(columns="month_no", inplace=True)
```

```python
df_Us_pivot
```

```python
top_3 = df_Us['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
```

```python
df_Us_pivot[top_3].plot(kind="line")
```

```python
df_US_Software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col="job_posted_month")
df_US_Software_pivot
```

```python
df_Us_pivot_merged = df_Us_pivot.merge(df_US_Software_pivot, on="job_posted_month")
df_Us_pivot_merged
```

```python
#we are using the sum function 
top_5 = df_Us_pivot_merged.sum().sort_values(ascending=False).head().index.to_list()
top_5
```

```python
df_Us_pivot_merged[top_5].plot(kind="line")
plt.title('Monthly job postings for tech jobs in the US')
plt.ylabel('Job Count')
plt.xlabel('2023')
plt.ylim(0,20000)
plt.show()
```
