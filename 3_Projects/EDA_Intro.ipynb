# Exploratory Data Analysis

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## Roles to Explore:
## Filtering for Data Analyst roles in the United States

```python
# Filtering for Data Analyst roles in the United States

df_DA_US = df[(df["job_title_short"] == "Data Analyst") & (df["job_country"] == "United States")]
df_DA_US
```

```python
# Counting the number of job location

df_DA_US ["job_location"].value_counts().head(10)

# We want to turn this into a dataframe before visualizing

df_plot = df_DA_US ["job_location"].value_counts().head(10).to_frame()

# set the theme
sns.set_theme(style="ticks")
# Creating a barchart visualization using seaborn
sns.barplot(data=df_plot, x="count", y="job_location", hue="count", palette="dark:b_r", legend=False)

# Customizing the visualization
plt.title("Counts of Job location for Data Analyst in the US")
plt.xlabel("Number of jobs")
plt.ylabel("")
plt.show ()

```

```python
## Creating a pie chart for benefit analysis of Data Analyst jobs in the US i.e (Requirements / Offerings)

#Creating a dict containing the columns in the data frame and their titles

dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}
dict_column

fig, ax =  plt.subplots(1,3, figsize=(15,5))

# The enumerate adds an index (i) to track loop iterations, dict_column.items() returns key-value pairs as tuples
for i , (column,title) in enumerate (dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), labels = ["False", "True"], autopct="%1.1f%%", startangle=90)
    ax[i].set_title(title)

# Formatting the visual
plt.suptitle('Benefit Analysis of Data Jobs', fontsize= 16)
plt.legend(loc='upper center', bbox_to_anchor=(-0.55, 0), ncol=2)


```

```python
# Counting the number of companies

df_DA_US ["company_name"].value_counts().head(10)

# We want to turn this into a dataframe before visualizing it

df_plot = df_DA_US ["company_name"].value_counts().head(10).to_frame()

# set the theme
sns.set_theme(style="ticks")
# Creating a barchart visualization using seaborn
sns.barplot(data=df_plot, x="count", y="company_name", hue="count", palette="dark:b_r", legend=False)

# Customizing the visualization
plt.title("Counts of Companies for Data Analyst in the US")
plt.xlabel("Number of jobs")
plt.ylabel("")
plt.show ()

#Note: Use cltrl Shif L to replace all same attributes at once.
```
