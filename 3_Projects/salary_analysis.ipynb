# How well do jobs and skills pay for Data Analysts?

**Methology**
1. Evaluate median salary for top 6 data jobs
2. Find median salary per skill for Data Analysts
3. Visualize for highest paying skills and most demanded skills

```python
#Importing Libraries

import ast
import seaborn as sns
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

#Data cleanup

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])

df["job_skills"] = df["job_skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_US = df[df['job_country'] == "United States"].copy().dropna(subset=['salary_year_avg'])
df_US
```

```python
job_titles = df_US['job_title_short'].value_counts().index[:6].tolist()

df_US_top6 = df_US[df_US['job_title_short'].isin (job_titles)]

df_US_top6

#order the job_titles by the median salary
job_order = df_US_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
job_order


```

```python
import seaborn as sns
sns.boxplot(data=df_US_top6, x = "salary_year_avg", y = "job_title_short", order=job_order)
sns.set_theme(style='ticks')

# plt.boxplot(job_list, labels=job_titles, vert=False)
plt.title('Salary Distribution in the United States')
plt.xlabel('Yearly Salary ($USD)')
plt.xlim(0, 600000) 
plt.show()
```

## Investigate Median Salary Vs Skill for Data Analysts

```python
df_DA_US = df[(df['job_country'] == "United States") & (df['job_title_short'] == "Data Analyst")].copy()

df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])

df_DA_US = df_DA_US.explode('job_skills')

df_DA_US[['salary_year_avg', 'job_skills']].head(5)
```

## Determine The Highest Paid Skills and Most Demanded Skills

```python
df_DA_US_top = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False)

df_DA_US_top = df_DA_US_top.head(10)
df_DA_US_top 
```

```python
df_DA_US_demand= df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)

df_DA_US_demand = df_DA_US_demand.head(10)
df_DA_US_demand 
```

```python
import seaborn as sns
fig, ax = plt.subplots(2,1)

sns.set_theme(style="ticks")


sns.barplot(data= df_DA_US_top, x = "median", y = df_DA_US_top.index, hue = "median", ax= ax[0], palette= "dark:b_r")

ax[0].legend().remove()
# top_pay[::-1].plot(kind='barh', ax= ax[0], y = "median", legend = False)
ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')
ax[0].set_ylabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

# high_demand[::-1].plot(kind='barh', ax= ax[1], y = "median", legend = False)

sns.barplot(data=df_DA_US_demand, x = "median", y = df_DA_US_demand.index, hue = "median", ax=ax[1], palette= "light:b")
ax[1].legend().remove()
ax[1].set_title('Top 10 Highest Demand Skills for Data Analysts')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary (USD)')

plt.tight_layout()
```
